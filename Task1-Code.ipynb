```python
import numpy as np
import pandas as pd
```

```python
uncleaned_Data = pd.read_csv("D:/ElevateLabs/Task1/marketing_campaign.csv", sep='\t')
```

```python
uncleaned_Data.info()
```

```python
# Checking the null/missing values
uncleaned_Data.isnull().sum()
```

```python
uncleaned_Data = uncleaned_Data.dropna(subset=["Income"])
```

```python
uncleaned_Data.isnull().sum()
```

```python
uncleaned_Data.duplicated().sum()
```

```python
uncleaned_Data.head()
```

```python
uncleaned_Data.dtypes
```

```python
uncleaned_Data['Education'] = uncleaned_Data['Education'].str.title().str.strip()
```

```python
uncleaned_Data['Marital_Status'] = uncleaned_Data['Marital_Status'].str.title().str.strip()
```

```python
uncleaned_Data['Dt_Customer'] = pd.to_datetime(uncleaned_Data['Dt_Customer'], dayfirst=True)
```

```python
uncleaned_Data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
categorical_cols = ['Education', 'Marital_Status']
for col in categorical_cols:
    uncleaned_Data[col] = uncleaned_Data[col].astype(str).str.lower().str.strip()  # clean text
    uncleaned_Data[col] = le.fit_transform(uncleaned_Data[col])
```

```python
uncleaned_Data.head()
```

```python
uncleaned_Data.to_csv("cleaned_marketing_campaign.csv", index=False)
```
